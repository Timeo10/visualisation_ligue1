```python
import pandas as pd
import folium
import re

# Chargement du CSV avec coordonnées
df = pd.read_csv("clubs_ligue1_2024_2025.csv")

# Nettoyage des capacités
df["Capacité"] = df["Capacité"].astype(str).str.replace(r"\s|\u202f", "", regex=True).str.replace(r"\[.*?\]", "", regex=True)
df["Capacité"] = pd.to_numeric(df["Capacité"], errors="coerce")

# Garde uniquement les lignes avec coordonnées valides
df_map = df.dropna(subset=["Latitude", "Longitude", "Capacité"])

# Création de la carte centrée sur la France
m = folium.Map(location=[46.5, 2.5], zoom_start=6)

# Marqueurs individuels (PAS DE CLUSTER)
for _, row in df_map.iterrows():
    popup_text = f"""
    <b>{row['Club']}</b><br>
    Stade : {row['Stade']}<br>
    Capacité : {int(row['Capacité'])} places<br>
    Entraîneur : {row['Entraîneur']}
    """
    folium.Marker(
        location=[row["Latitude"], row["Longitude"]],
        popup=popup_text
    ).add_to(m)

# Affiche la carte dans le notebook
m
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Charger les données
df = pd.read_csv("clubs_ligue1_2024_2025.csv")

# Nettoyage des données
df["Capacité"] = df["Capacité"].astype(str).str.replace(r"\s|\u202f", "", regex=True).str.replace(r"\[.*?\]", "", regex=True)
df["Capacité"] = pd.to_numeric(df["Capacité"], errors="coerce")

df["Classement (dernier disponible)"] = df["Classement (dernier disponible)"].astype(str).str.extract(r'(\d+)')
df["Classement (dernier disponible)"] = pd.to_numeric(df["Classement (dernier disponible)"], errors="coerce")

# Graphique 1 : Capacités des stades
plt.figure(figsize=(12, 6))
df_sorted = df.sort_values(by="Capacité", ascending=False)
plt.bar(df_sorted["Club"], df_sorted["Capacité"])
plt.xticks(rotation=45, ha='right')
plt.title("Capacité des stades de Ligue 1")
plt.ylabel("Capacité")
plt.tight_layout()
plt.show()

# Graphique 2 : Classement (si disponible)
df_ranked = df.dropna(subset=["Classement (dernier disponible)"]).sort_values(by="Classement (dernier disponible)")
plt.figure(figsize=(12, 6))
plt.bar(df_ranked["Club"], df_ranked["Classement (dernier disponible)"])
plt.xticks(rotation=45, ha='right')
plt.title("Classement des clubs (dernier connu)")
plt.ylabel("Position")
plt.gca().invert_yaxis()  # 1er en haut
plt.tight_layout()
plt.show()

# Graphique 3 : Budgets (si colonne présente)
if "Budget (€M)" in df.columns:
    df["Budget (€M)"] = df["Budget (€M)"].astype(str).str.replace(",", ".").str.extract(r'(\d+\.?\d*)')
    df["Budget (€M)"] = pd.to_numeric(df["Budget (€M)"], errors="coerce")

    df_budget = df.dropna(subset=["Budget (€M)"]).sort_values(by="Budget (€M)", ascending=False)
    plt.figure(figsize=(12, 6))
    plt.bar(df_budget["Club"], df_budget["Budget (€M)"])
    plt.xticks(rotation=45, ha='right')
    plt.title("Budgets des clubs de Ligue 1 (€ millions)")
    plt.ylabel("Budget (€M)")
    plt.tight_layout()
    plt.show()
```
